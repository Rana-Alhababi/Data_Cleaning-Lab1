### Do the following steps for each column with missing values.

* Identity the column

```python
#Identity the column

columns_with_missing_values = df.columns[df.isnull().any()]
print("Columns with missing values:")
print(columns_with_missing_values)
```

```python
#How much precentage the missing value represents
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("Missing value percentages:")
print(missing_percentage)
```

```python
# Calculate the overall sum of the percentages of missing values
overall_missing_percentage = missing_percentage.sum()

print("Overall sum of missing value percentages:", overall_missing_percentage)
```

###Since it's a categorical column representing the names of employees. Instead of dropping or imputing values, one option is to assign a placeholder value such as Unknown

```python
# Replace missing values with "Unknown"
df['Employee_Name'].fillna("Unknown", inplace=True)

# Print the updated column with missing values handled
print("Data with missing values handled in 'Employee_Name' column:")
df['Employee_Name']
```

###Since It's a numerical column the best way is  Imputation with Mean/Median: One approach is to impute the missing values with the mean or median of the 'Salary' column. This method assumes that the missing values are missing at random and that the distribution of salaries is approximately normal.

```python
# Calculate the mean or median of the 'Salary' column
mean_salary = df['Salary'].mean()
#OR median_salary = df['Salary'].median()

# Impute missing values with the mean or median
df['Salary'].fillna(mean_salary, inplace=True)
df['Salary']
```

### Since It's Date of Birth I was confused, but I applied two methods First is Imputation with Mean/Median: One approach is to impute the missing values with the mean or median of the available dates of birth. However, imputing dates may not be appropriate in all cases, especially if the missing values are not missing at random. Second is Create a Missing Indicator: Instead of imputing the missing values, you can create a separate column that indicates whether the 'DOB' value was missing or not. This approach allows you to retain the information that a value was missing without distorting the distribution of dates of birth because sometime we can not We predict a person's date of birth


```python
# Calculate the mode of the 'DOB' column
mode_dob = df['DOB'].mode()[0]

# Impute missing values with the mode
df['DOB'].fillna(mode_dob, inplace=True)
df['DOB']

#____________________OR___________________________

#  Create a missing indicator column
df['DOB_Missing'] = df['DOB'].isnull().astype(int)
df['DOB']
```

### Since It's sex I was confused, but I applied two methods Mode Imputation: If the missing values in the 'Sex' column are relatively few, you can impute them with the mode (the most frequent value) of the 'Sex' column. This assumes that the missing values are missing at random and that the mode represents a reasonable estimate for the missing values.or we Create a Missing Indicator: Instead of imputing the missing values, you can create a separate column that indicates whether the 'Sex' value was missing or not. This approach allows you to retain the information that a value was missing without distorting the distribution of the existing categories.

```python
# Calculate the mode of the 'Sex' column
mode_sex = df['Sex'].mode()[0]

# Impute missing values with the mode
df['Sex'].fillna(mode_sex, inplace=True)
df['Sex']

#____________________OR___________________________


# Create a missing indicator column
df['Sex_Missing'] = df['Sex'].isnull().astype(int)
```

### Since It's DateofTermination I was confused, but I applied two methods Imputation with a Specific Value: If the missing values in the 'DateofTermination' column indicate that the employee has not been terminated, you can impute the missing values with a specific value, such as 'Still Employed' or 'N/A'. or Create a Termination Indicator: Instead of imputing the missing values, you can create a separate column that indicates whether the employee has been terminated or not. This approach allows you to retain the information that a termination date was missing without distorting the distribution of termination dates.

```python

# Impute missing values with a specific value
df['DateofTermination'].fillna('Still Employed', inplace=True)


#____________________OR___________________________

# Create a termination indicator column
df['Terminated'] = df['DateofTermination'].notnull().astype(int)
```

### Since It's managerID I was confused, but I applied two methods Imputation with a Specific Value: If the missing values in the 'ManagerID' column represent a specific category (e.g., no manager assigned), you can impute the missing values with a specific value, such as 'No Manager' or 'N/A'. or Imputation with Mode: If the missing values in the 'ManagerID' column are relatively few and there is a dominant 'ManagerID' value (mode), you can impute the missing values with the mode of the 'ManagerID' column.

```python
# Impute missing values with a specific value
df['ManagerID'].fillna('No Manager', inplace=True)


#____________________OR___________________________


# Calculate the mode of the 'ManagerID' column
mode_manager_id = df['ManagerID'].mode()[0]

# Impute missing values with the mode
df['ManagerID'].fillna(mode_manager_id, inplace=True)
```

```python
# Calculate the mean or median of the 'Absences' column
mean_absences = df['Absences'].mean()
# median_absences = df['Absences'].median()

# Impute missing values with the mean or median
df['Absences'].fillna(mean_absences, inplace=True)
```
